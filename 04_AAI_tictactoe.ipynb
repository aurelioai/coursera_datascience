```python
# # TICTACTOE PROGRAM #
#####################

#FUNCTION DEFINITION

# Import clear output function
from IPython.display import clear_output

# Welcome message and instructions
def welcome():
    print('Welcome to TicTacToe!')
    print('Two players required. Player 1 goes first.')
    print('When prompted, choose a number to replace it on the board with your mark.')
    print('Win if you get three slots with your mark aligned.')
# Player1 chooses its mark
def playermark():
    valid = False
    while not valid:
        play1mark = input('Player 1, choose your mark (X or O): ').upper()
        if play1mark=='X':
            play2mark = 'O'
            valid = True
        elif play1mark=='O':
            play2mark = 'X'
            valid = True
        else:
            print('Please enter a valid mark (X or O).')
    return play1mark,play2mark
# The board
def myboard(row1,row2,row3):
    myhline = '------------'
    print(' '+row1[0]+' | '+row1[1]+' | '+row1[2])
    print(myhline)
    print(' '+row2[0]+' | '+row2[1]+' | '+row2[2])
    print(myhline)
    print(' '+row3[0]+' | '+row3[1]+' | '+row3[2])
# Player input (do not forget validation)
def playinput(player,mark,taken):
    valid = False
    while not valid:
        choice = input('Player {}, choose a slot to fill with {}'.format(player,mark)+': ')
        if choice.isdigit():
            if int(choice) in range(1,10):
                if int(choice) not in taken:
                    valid = True
                else:
                    print('The slot you chose is already taken.')
            else:
                print('The slot you chose is not in the board.')
        else:
            print('You did not write a number.')
    taken.append(int(choice))
    return int(choice),taken
# Update rows
def updaterow(row1,row2,row3,mark,choice,pset):
    if choice in range(1,4):
        row1[choice-1] = mark
        pset.add(choice)
    elif choice in range(4,7):
        row2[choice-4] = mark
        pset.add(choice)
    else:
        row3[choice-7] = mark
        pset.add(choice)
def checkwin(playerset):
    winners=[{1,2,3},{4,5,6},{7,8,9},
             {1,4,7},{2,5,8},{3,6,9},
             {1,5,9},{3,5,7}]
    for winset in winners:
        if winset.issubset(playerset):
            return True
    return False
def playagain():
    valid = False
    while not valid:
        choice = input('Do you want to play? Y/N: ')
        if choice.upper()=='Y':
            clear_output()
            return True
        if choice.upper()=='N':
            return False
        else:
            print('Please choose a valid option')
####################################################################
####################################################################
####################################################################
# GAME INITIALIZATION
# Welcome message and instructions
clear_output()
welcome()
gameon = True
while gameon:
    r1=[' ',' ',' ']
    r2=[' ',' ',' ']
    r3=[' ',' ',' ']
# Player1 chooses its mark
    p1m,p2m = playermark()
    taken=[]
    p1set = set()
    p2set = set()
# Show the board slots
    clear_output()
    print('This is the board. Remember the numbers for your marks!')
    r1=['1','2','3']
    r2=['4','5','6']
    r3=['7','8','9']
    myboard(r1,r2,r3)
    r1=[' ',' ',' ']
    r2=[' ',' ',' ']
    r3=[' ',' ',' ']

# LET`S PLAY
    while True:
        choice,taken = playinput(1,p1m,taken)
        clear_output()
        updaterow(r1,r2,r3,p1m,choice,p1set)
        myboard(r1,r2,r3)
        if checkwin(p1set):
            message = 'Congratulations, Player 1. You are the winner!'
            break
        if len(taken)==9:
            message = 'The board is full with no winner.'
            break
        choice,taken = playinput(2,p2m,taken)
        clear_output()
        updaterow(r1,r2,r3,p2m,choice,p2set)
        myboard(r1,r2,r3)
        if checkwin(p2set):
            message = 'Congratulations, Player 2. You are the winner!'
            break
#Â Winner!
    print(message)
    gameon = playagain()
print('\nThank you for playing! Come again soon!')
```
