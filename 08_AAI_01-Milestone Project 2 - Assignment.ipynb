```python
import random
suits = ("Corazones", "Espadas", "Tréboles", "Diamantes")
ranks = ('Dos', 'Tres', 'Cuatro', 'Cinco', 'Seis', 'Siete', 'Ocho', 'Nueve', 'Diez', 'Jack', 'Reina', 'Rey', 'As')
values = {'Dos': 2, 'Tres': 3, 'Cuatro': 4, 'Cinco': 5, 'Seis': 6, 'Siete': 7, 'Ocho': 8, 'Nueve': 9, 'Diez': 10,
          'Jack': 10, 'Reina': 10, 'Rey': 10, 'As': (1, 11)}
```

```python
class Card:

    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit
        self.value = values[rank]

    def __str__(self):
        return self.rank + " de " + self.suit


class Deck:

    def __init__(self):

        self.allcards = []
        for suit in suits:
            for rank in ranks:
                self.allcards.append(Card(rank, suit))

    def __str__(self):
        return "Esta baraja tiene {} cartas.".format(len(self.allcards))

    def shuffle(self):
        random.shuffle(self.allcards)

    def printdeck(self):
        for i in range(len(self.allcards)):
            print(self.allcards[i])

    def dealcard(self):
        return self.allcards.pop(0)
```

```python
class Player:

    def __init__(self, name):

        self.name = name
        self.wallet = 100
        self.hand = []
        self.handtotal = 0
        print("{} comienza con una cartera de ${}.".format(self.name, self.wallet))

    def __str__(self):

        return "{} tiene ${} en la cartera y {} cartas en la mano.".format(self.name, self.wallet, len(self.hand))

    def addcard(self, cards):

        if (type(cards) == type([])):
# If "cards" is a list, add the whole list to my hand...
            self.hand.extend(cards)
        else:
# ... otherwise, "cards" is only one element. Add it with "append".
            self.hand.append(cards)
        self.handtotal = 0
# Calculate the total of points.
        for card in self.hand:
# Once an "As" has a value it does not change. ## Should I change this??
            if card.rank=="As":
                if (self.handtotal+11>21):
                    self.handtotal += 1
                else:
                    self.handtotal += 11
            else:
                self.handtotal += card.value
        return None

    def bet(self, amount):

        accepted = False
        final = self.wallet - amount
        if (final < 0):
            print("{} puede apostar hasta ${}.".format(self.name, self.wallet))
            accepted = False
        else:
            self.wallet = final
            print("{} apuesta ${}, le quedan ${} en la cartera.".format(self.name, amount, self.wallet))
            accepted = True
        return accepted

    def updt_wallet(self, amount):

        self.wallet += amount
        return print('{} tiene ahora ${} en la cartera.'.format(self.name, self.wallet))

        
    def newgame(self):
        self.hand = []
        self.handtotal = 0
        return None


class Dealer:

    def __init__(self):

        self.hand = []
        self.handtotal = 0

    def addcard(self, cards):

        if (type(cards) == type([])):
            self.hand.extend(cards)
        else:
            self.hand.append(cards)
# Calculate the total of points.
        self.handtotal = 0
        for card in self.hand:
# Once an "As" has a value it does not change. ## Should I change this??
            if card.rank=="As":
                if (self.handtotal+11>21):
                    self.handtotal += 1
                else:
                    self.handtotal += 11
            else:
                self.handtotal += card.value
        return None
    
    def newgame(self):
        self.hand = []
        self.handtotal = 0
        return None
```
